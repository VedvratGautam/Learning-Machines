# Pandas Basics

Goal of this notebook is to understand Panda methods and working for Data preprocessing and EDA

Importing Dependencies

```python
import pandas as pd
```

```python
#Reading the csv

df = pd.read_csv('/content/world_population.csv')
```

```python
df
```

## Filtering

```python
df
```

```python
df[df['Rank'] <= 10] #Filtering values in rank column <= 10
```

```python
specific_countries = ['Bangladesh' , 'Brazil' , 'India'] #Getting row Data for specific countries

df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('United')] #Checking Country column for strings with 'United'
```

```python
df2 = df.set_index('Country') #Changing Index to country(sorted alphabetically by default)
```

```python
df2
```

```python
df2.filter(items = ['Continent' , 'CCA3'], axis = 1) #Filtering which column contains the items
```

```python
df2.filter(items = ['Zimbabwe'], axis = 0) #Filtering which row contains the items
```

```python
df2.filter(like = 'Unite', axis = 0)  #Filtering which row contains the items
```

```python
df2.loc['China'] #Getting all data for China
```

```python
df2.iloc[2] #Searching within index
```

Order By

```python
df
```

```python
df[df['Rank'] < 10].sort_values(by = 'Rank') #Sorting rank values <10 in ascending order
```

```python
df[df['Rank'] < 10].sort_values(by = ['Country' , 'Rank'] , ascending=[False, True]) #Sorting by multiple features
```
