# Visualisations using Pandas

```python
#Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Ice Cream Ratings.csv')
```

```python
df
```

```python
df = df.set_index('Date')
```

```python
df.describe()
```

```python
df.mean(numeric_only= True)
```

## Styles

```python
print(plt.style.available)
plt.style.use('Solarize_Light2')
```

```python
df.plot(kind = 'line')
```

```python
df.plot(subplots = True)
```

```python
df.plot(subplots = True, xlabel = 'Dates', ylabel = 'Daily Ratings')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'barh' , stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar')
```

```python
df.plot.scatter(x = 'Flavor Rating' , y = 'Texture Rating', s=30, c = 'red')
```

```python
df.plot.hist(bins = 20, )
```

```python
df.plot.box()
```
