# Data Preprocessing using Pandas

I will attempt to 'clean' a messy dataset in order to make the dataset ready for EDA.

```python
#Importing Libraries

import pandas as pd
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
```

```python
df
```

```python
df.describe()
```

## Removing Duplicates

```python
df = df.drop_duplicates()
```

```python
df
```

## Removing 'Not_Useful_Column' Column

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
df.isna().sum()
```

## Working on Last_Name column

```python
#Stripping typos

df['Last_Name'] = df['Last_Name'].str.strip('123./_')

df
```

## Cleaning Phone Numbers

```python
#Writing a regex to clean up 'Phone_Number' column

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]' , '' , regex = True)
df['Phone_Number']
```

```python
#Adding '-' after 3 digits of a phone number to formatt it correctly

df['Phone_Number'] = df['Phone_Number'].apply(lambda number: str(number)) #Converting values to strings
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda number: number[0:3] + '-' + number[3:6] + '-' + number[6:]) #Adding the -
```

```python
df
```

```python
#Removing nan-- and Na-- values

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--' , '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--' , '')
```

```python
df
```

## Address

```python
#Splitting address into its components
df[['Street_Address', 'State' , 'Zip_Code']] = df['Address'].str.split(',',n = 2, expand = True)
```

```python
df
```

## Paying Customer

```python
#Changing all No's to N and all Yes's to Y

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

```

```python
df
```

## N/A values

```python
df = df.fillna('')
df
```

## Further cleaning Data

Instruction - Only provide details for people who wish to be contacted

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes' , 'Y')

df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No' , 'N')

df
```

```python
indices_to_drop = []

for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    indices_to_drop.append(x)

df.drop(indices_to_drop, inplace=True)

df
```

```python
#Dropping empty phone number

phone_numbers_to_drop = []

for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    phone_numbers_to_drop.append(x)

df.drop(phone_numbers_to_drop, inplace=True)
```

```python
df
```

```python
#Resetting index

df.reset_index(drop=True)
```
