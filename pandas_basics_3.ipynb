#Working with multiple Dataframes

Merge, Join and Concatenate

```python
#Dependencies

import pandas as pd
```

```python
#Loading the data

df1 = pd.read_csv('/content/LOTR.csv')
```

```python
df1
```

```python

df2 = pd.read_csv('/content/LOTR 2.csv')
```

```python
df2
```

## Merge - Outer, Inner, Left, Right

```python
#Inner join is default --> Merges common features

df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName']) #df1 is left df by syntax
```

```python
#Outer join --> Merges all features

df1.merge(df2, how = 'outer')
```

```python
#Left Join --> Copies left df and then tries to merge common features

df1.merge(df2, how = 'left')
```

```python
#Right Join --> Copies right df and then tries to merge common features

df1.merge(df2, how = 'right')
```

```python
#Cross Join --> Gives cross product of the two dataframes (rows)

df1.merge(df2 , how = 'cross')
```

## Joins

```python
#Commonly used when working with indices

df1.join(df2,  lsuffix = '_left', rsuffix = '_right')
```

```python
df4 = df1.set_index('FellowshipID')
```

```python
df4.join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')
```

## Concatnate

Works like regular concat

```python
pd.concat([df1 , df2]) #Default is a outer join
```
