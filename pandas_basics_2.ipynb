# Learning groupby and aggregate functions in Pandas

Aggregate functions add rows together

```python
#Dependencies

import pandas as pd
```

```python
#Loading the Data

df = pd.read_csv('/content/Flavors.csv')
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor') #Groups rows by Base Flavour
```

```python
group_by_frame.mean(numeric_only= True)
```

```python
group_by_frame.count() #Get counts of rows that were aggregated
```

```python
group_by_frame.min() #For non numeric columns sorts alphabetically
```

```python
group_by_frame.max(numeric_only= True)
```

```python
group_by_frame.sum()
```

```python
group_by_frame.sum(numeric_only= True)
```

```python
#Need to pass dict to aggregate function

group_by_frame.agg({'Flavor Rating': ['mean' , 'max' , 'count' , 'sum'], 'Texture Rating': ['mean' , 'max' , 'count' , 'sum']})
```

groupby - Multiple columns

```python
group_by_multiple = df.groupby(['Base Flavor' , 'Liked'])
```

```python
group_by_multiple.mean(numeric_only=True)
```

```python
group_by_multiple.agg({'Flavor Rating': ['mean' , 'max' , 'count' , 'sum'], 'Texture Rating': ['mean' , 'max' , 'count' , 'sum']})
```

```python
group_by_multiple.describe()
```
